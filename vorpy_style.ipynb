```python
def load_player_data(csv_file, rank_col_str, pos_col_str, value_col_str, float_cols, rep_level):
    
    with open(csv_file) as csvfile:
        reader = csv.reader(csvfile)

        player_data = []
        headers = []
        row_idx = 0
        
        for row in reader: 
            if row_idx == 0:
                col_idx = 0
                name_idx = 0

                # build the headers list
                for col in row:
                    headers.append(row[col_idx])
                    col_idx += 1
                headers.append("Vorp")

            else:
                col_idx = 0
                player_list = []

                # build the player data dict
                for col in row:
                    cell_val = float(row[col_idx]) if col_idx in float_cols else row[col_idx]
                    if col_idx == idx_from_header(pos_col_str,headers): cell_val = row[col_idx].split()
                    player_list.append(cell_val)
                    col_idx += 1
                player_list.append(0.0)

                player_data.append(player_list)

            row_idx+=1   
        
    batters = ['C', 'SS', 'OF', '1B', '2B', '3B']
    pitchers = ['RP','SP']
    starting_pos = batters + pitchers
    
    # Calculate replacement values
    replace_values = {pos:calc_replace_val_by_pos(rep_level,pos,idx_from_header(rank_col_str,headers),\
                                                         idx_from_header(pos_col_str,headers),\
                                                         idx_from_header(value_col_str,headers),\
                                                         player_data) for pos in starting_pos}


    # Calculate VORP
    for i, player in enumerate(player_data):
        player_pos = player_data[i][idx_from_header(pos_col_str,headers)]
        vorp_list = []
        for pos in player_pos:
            vorp_list.append(player_data[i][idx_from_header(value_col_str,headers)] - replace_values[pos])
        player_data[i][idx_from_header("Vorp",headers)] = max(vorp_list)
        
    return player_data, headers

def calc_replace_val_by_pos(replacement_level, player_pos, rank_col, pos_col, value_col, data):
    avg_val = 0.0
    return sum(find_next_three_replacement_values(replacement_level, player_pos, rank_col, pos_col, value_col, data))/3.0


def find_next_three_replacement_values(replacement_level, player_pos, rank_col, pos_col, value_col, data):
    ret = []
    level = replacement_level
    
    for i in range(3):
        idx, val = find_next_replacement_value(level, player_pos, rank_col, pos_col, value_col, data)
        ret.append(val)
        level = idx + 1
        
    return ret

def find_next_replacement_value(replacement_level, player_pos, rank_col, pos_col, value_col, data):
    for player_data in data:
        
        pos   = player_data[pos_col]
        rank  = player_data[rank_col]
        value = player_data[value_col]
                
        if rank > replacement_level and player_pos in pos: return rank, value

def idx_from_header(idx_value, header):
    i = 0
    for val in header:
        if idx_value == val: return i
        i+=1
```
